```python
import pandas as pd

df = pd.read_csv('train.csv')
df
```

```python
df[df['Age'].isnull() == True]
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
df
```

```python
df[df['Embarked'].isnull()]
```

```python
df['Embarked'].value_counts()
```

```python
df['Embarked']=df['Embarked'].fillna('S')
df.loc[[61,829], :] #顯示列索引61,829的資料
```

```python
df = df.drop('Cabin', axis=1)
df.info()
```

```python
df[df.duplicated()]
```

```python
s={'female':0, 'male':1}
df['Sex']=df['Sex'].map(s)
e={'S':0, 'C':1, 'Q':2}
df['Embarked']=df['Embarked'].map(e)
df.head()
```

```python
df['Survived'].value_counts()
```

```python
df['Survived'].value_counts().plot(kind='pie', autopct='%1.2f%%')
```

```python
df['Sex'].value_counts()
```

```python
df['Sex'].value_counts().plot(kind='pie',autopct='%1.2f%%')
```

```python
df['Pclass'].value_counts()
```

```python
df['Pclass'].value_counts().plot(kind='pie',autopct='%1.2f%%')
```

```python
#女、男乘客的人數
df.groupby(['Sex'])['PassengerId'].count()
```

```python
df.groupby(['Sex','Survived'])['PassengerId'].count().plot(kind='bar', rot=1)
```

```python
#不同性別生還人數/不同性別人數
ss = df.groupby(['Sex','Survived'])['PassengerId'].count() / \
     df.groupby(['Sex'])['PassengerId'].count() * 100
ss
```

```python
ss.plot(kind='bar', color=['r','g'], rot=0)
```

```python
#三種艙等的生還和死亡人數
df.groupby(['Pclass','Survived'])['PassengerId'].count()
```

```python
df.groupby(['Pclass','Survived'])['PassengerId'].count().plot(kind='bar', rot=0)
```

```python
#不同艙等生還人數/不同艙等人數
ps = df.groupby(['Pclass','Survived'])['PassengerId'].count() / \
     df.groupby(['Pclass'])['PassengerId'].count() * 100
ps
```

```python
ps.plot(kind='bar', rot=0)
```

```python
df.plot(kind='scatter', x='Age', y='Fare', alpha=0.5)
```

```python
# c='Survived' 代表根據生還狀態上色
# colormap 可以選擇顏色主題，'RdYlGn' 是 紅-黃-綠
df.plot(kind='scatter', x='Age', y='Fare', c='Survived', colormap='RdYlGn', alpha=0.6)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# hue='Survived' 會自動幫你根據生還與否區分顏色
sns.scatterplot(data=df, x='Age', y='Fare', hue='Survived', palette='viridis')

plt.title('Age vs Fare (Survival Analysis)')
plt.show()
```

# 02：鳶尾花

```python
import pandas as pd
df = pd.read_csv('Iris.csv')
df
```

```python
df=df.drop('Id', axis=1)
df.head()
```

```python
df.head()
```

```python
df.info()
```

```python
df[df.duplicated()]
```

```python
df = df.drop_duplicates()
df[df.duplicated()]
```

```python
df.reset_index(drop=True)
```

```python
s = {'Iris-setosa':0, 'Iris-versicolor':1, 'Iris-virginica':2 }
df['Species']=df['Species'].map(s)
df.head()
```

```python
df.head()
```

```python
c = {0:'r', 1:'g', 2:'b'}
df['colors'] = df['Species'].map(c)
df
```

```python
df.plot(kind='scatter', x='SepalLengthCm', y='Species', c=df['colors'])
```

```python
df.plot(kind='scatter', x='SepalWidthCm', y='Species', c=df['colors'])
```

```python
df.plot(kind='scatter', x='PetalLengthCm', y='Species', c=df['colors'])
```

```python
df.plot(kind='scatter', x='PetalWidthCm', y='Species', c=df['colors'])
```
